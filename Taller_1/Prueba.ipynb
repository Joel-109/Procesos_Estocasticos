```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("Base.xlsx")
df.head()

df = df.drop(columns=['Fecha'])
df.head()
```

```python

for column in df.columns:

    df[column] = df[column].str.replace(".","",regex=False)
    df[column] = df[column].str.replace(",",".",regex=False)

    df[column] = df[column].astype(float)

df.head()
```
